```python
import graph_learning_gaussian as glg
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import scipy.io
```

```python
# Noisy signals over Erdos Reyni graph from main_gaussian_new.m 
# and parameters used.
X_noisy = scipy.io.loadmat('test_signals/X_noisy.mat')['X_noisy']
param = {'N':20, 'max_iter':50, 'alpha':0.01, 'beta':0.631}
```

```python
L, Y = glg.graph_learning_gaussian(X_noisy, param)
```

```python
# Learned laplacian from matlab implementation
ML = scipy.io.loadmat('test_signals/L0.mat')['L_0']
```

```python
Gt = nx.from_numpy_matrix(ML)
Et = nx.from_numpy_matrix(L)
```

```python
for i in range(len(L)):
    for j in range(len(L[i])):
        if abs(L[i][j]) < .1:
            L[i][j] = 0
```

```python
im1 = plt.imshow(L, cmap='jet', interpolation='nearest')
plt.colorbar()
plt.show()
plt.savefig("est_hm.png")
```

```python
im1 = plt.imshow(ML, cmap='jet', interpolation='nearest')
plt.colorbar()
plt.show()
plt.savefig("GT_hm.png")
```

```python
est_G = nx.from_numpy_matrix(np.diag(np.diag(L)) - L)
edges,weights = zip(*nx.get_edge_attributes(est_G,'weight').items())
nodes = est_G.nodes()
node_color = ['black']*len(nodes)
size = [w*5 for w in weights]
pos = nx.circular_layout(est_G)
ec = nx.draw_networkx_edges(est_G, pos, 
                            edgelist=edges,
                            edge_cmap=plt.cm.jet)
nc = nx.draw_networkx_nodes(est_G,pos,
                            nodelist=nodes,
                            node_color=node_color, 
                            node_size = 100,
                            alpha = .9)
plt.savefig("est_graph.png")
```

```python
est_G = nx.from_numpy_matrix(np.diag(np.diag(ML)) - ML)
edges,weights = zip(*nx.get_edge_attributes(est_G,'weight').items())
nodes = est_G.nodes()
node_color = ['black']*len(nodes)
size = [w*5 for w in weights]
pos = nx.circular_layout(est_G)
ec = nx.draw_networkx_edges(est_G, pos, 
                            edgelist=edges, 
                            edge_color=weights, 
                            width=size,
                            edge_cmap=plt.cm.jet)
nc = nx.draw_networkx_nodes(est_G,pos,
                            nodelist=nodes,
                            node_color=node_color, 
                            node_size = 100,
                            alpha = .9)
plt.savefig("gt_graph.png")
```
